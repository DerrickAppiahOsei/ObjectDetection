```python
####Problem 1

import numpy as np
import matplotlib.pyplot as plt
# Define the displacement functions u1(x) and u2(x) as provided
def u1(x):
    return (1 - x / 100) * 0 + (x / 100) * 0.004233

def u2(x):
    return (1 - (x - 80) / 80) * 0.004233 + ((x - 80) / 80) * 0.00441256

# Define the ranges for u1 and u2
x1_extended = np.linspace(0, 100, 300)
x2_extended = np.linspace(100, 180, 300)

# Compute the function values for the displacements
y_u1 = u1(x1_extended)
y_u2 = u2(x2_extended)

# Plotting the displacement functions
plt.figure(figsize=(10, 6))
plt.plot(x1_extended, y_u1, label='Element 1 $u^{(1)}(x)$')
plt.plot(x2_extended, y_u2, label='Element 2 $u^{(2)}(x)$')
plt.xlabel('x (mm)')
plt.ylabel('Displacement (m)')
plt.title('Displacement Functions $u^{(1)}(x)$ and $u^{(2)}(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the stress values from the document
sigma1 = 42.33  # N/cm²
sigma2 = 2.62  # N/cm²

# Create arrays filled with these constant stress values over their respective x ranges
y_sigma1_new = np.full_like(x1_extended, sigma1)
y_sigma2_new = np.full_like(x2_extended, sigma2)

# Plotting the stresses
plt.figure(figsize=(10, 6))
plt.plot(x1_extended, y_sigma1_new, label='Element 1 $\sigma^{(1)}(x)$ (N/cm²)')
plt.plot(x2_extended, y_sigma2_new, label='Element 2 $\sigma^{(2)}(x)$ (N/cm²)')
plt.xlabel('x (cm)')
plt.ylabel('Stress (N/cm²)')
plt.title('Stress Distribution $\sigma^{(1)}(x)$ and $\sigma^{(2)}(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
