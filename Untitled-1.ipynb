```python
import numpy as np
import matplotlib.pyplot as plt
# Load the .npz file to see the contents
file_path = '/home/m3-learning/Documents/myML/ValidationData-20231107T163458Z-001/ValidationData/Stack007.npz'
data = np.load(file_path)
data_files = list(data.files)

data_files
```

```python
# Adjusting the approach to handle 0-dimensional arrays
array_details_corrected = {}
for key in data_files:
    array_info = {
        "shape": data[key].shape,
    }
    # Handle 0-dimensional arrays
    if data[key].ndim == 0:
        array_info["value"] = data[key].item()
    else:
        array_info["preview"] = data[key][:5]
    
    array_details_corrected[key] = array_info

array_details_corrected
```

```python
# Load the initial .npz file again to access X and y arrays
data_initial = np.load('/home/m3-learning/Documents/myML/ValidationData-20231107T163458Z-001/ValidationData/Stack007.npz')

# Extract the first image from both X and y for visualization
image_X = data_initial['X'][0, 0]
image_y = data_initial['y'][0, 0]

# Plotting both images side by side for comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display the image from X
axes[0].imshow(image_X, cmap='gray')
axes[0].set_title('Image from X')
axes[0].axis('off')  # Hide axes for better visibility

# Display the corresponding label image from y
axes[1].imshow(image_y, cmap='gray')
axes[1].set_title('Label Image from y')
axes[1].axis('off')  # Hide axes for better visibility

plt.show()

```

```python
# Load the results file to see the contents
results_file_path = '/home/m3-learning/Documents/myML/ValidationResults/Stack007_result.npz'
results_data = np.load(results_file_path)
results_files = list(results_data.files)

# Plotting MAE, Recall, and Precision
# Adjusting the aspect ratio to stretch the heatmaps horizontally
fig, axes = plt.subplots(1, 3, figsize=(20, 6), gridspec_kw={'width_ratios': [1, 1, 1], 'wspace': 0.5})

# Mean Absolute Error
mae = results_data['mae']
im_mae = axes[0].imshow(mae, cmap='viridis', aspect='auto')  # Set aspect='auto' to stretch
axes[0].set_title('Mean Absolute Error (MAE)')
axes[0].set_xlabel('Experiment')
axes[0].set_ylabel('Sample')
fig.colorbar(im_mae, ax=axes[0])

# Recall
recall = results_data['recall']
im_recall = axes[1].imshow(recall, cmap='viridis', aspect='auto')
axes[1].set_title('Recall')
axes[1].set_xlabel('Experiment')
axes[1].set_ylabel('Sample')
fig.colorbar(im_recall, ax=axes[1])

# Precision
precision = results_data['precision']
im_precision = axes[2].imshow(precision, cmap='viridis', aspect='auto')
axes[2].set_title('Precision')
axes[2].set_xlabel('Experiment')
axes[2].set_ylabel('Sample')
fig.colorbar(im_precision, ax=axes[2])

plt.tight_layout()
plt.show()


```
